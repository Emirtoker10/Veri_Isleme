```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Dell/Desktop/Yazılım/Jupyter Notebook/Veriler/e_ticaret_kullanicilari.csv")
df
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df
```

```python
df["signup_date"] = pd.to_datetime(df["signup_date"],errors = "coerce") # string tarihleri datetime'a çevirdik artık istediğimiz gibi
df                                                                      # gün ay yıl verilerine ulaşabileceğiz
```

```python
df["signup_date"] = df["signup_date"].fillna(df["signup_date"].median())
#df["signup_date"] = df["signup_date"].fillna(pd.Timestamp("2023-01-01")) belirli bir tarihle doldurma
df
```

```python
df["email_verified"] = df["email_verified"].map({True:1,False:0})
df
```

```python
df["segment"] = df["segment"].map({"bronze":1,"silver":2,"gold":3})
df
```

```python
def categorize_purchase(x):
    if x == 0:
        return "no_purchase"
    elif x < 3:
        return "low"
    elif 4<= x <=10:
        return "medium"
    else:
        return "high"

df["purchase_level"] = df["purchase_count"].apply(categorize_purchase)
df
```

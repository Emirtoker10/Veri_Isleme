```python
from sklearn.linear_model import LinearRegression 
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Dell/Desktop/Yazılım/Jupyter Notebook/Veriler/Audi_A1_listings.csv")
df.head()
```

```python
df.drop(columns = ["index","href","MileageRank","PriceRank","PPYRank","Score","Type","PPY"],inplace=True)
df
```

```python
df.columns = ["Yil","Mil","Motor","PS","Vites","Yakit","Sahip","Fiyat(£)"]
df
```

```python
df["Motor"] = df["Motor"].str.replace("L","")
df["Motor"] = pd.to_numeric(df["Motor"])
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns = ["Vites","Yakit"],drop_first=True)
df.head()
```

```python
df[["Vites_Manual", "Yakit_Petrol"]] = df[["Vites_Manual", "Yakit_Petrol"]].astype(int) # get_dummies ile true falseye dönüştürdüğümüz sütunları
df                                                                                      # 0 ve 1 lere dönüştürdük
```

```python
df.columns
```

```python
y = df["Fiyat(£)"]
x = df.drop(["Fiyat(£)"],axis=1)
```

```python
l = LinearRegression()
model = l.fit(x,y)
```

```python
model.predict([[2017,30000,1.6,110,1,0,1]])
```

```python
model.score(x,y)
```

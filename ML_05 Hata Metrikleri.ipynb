```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df = pd.read_csv("C:/Users/Dell/Desktop/Yazılım/Jupyter Notebook/Veriler/insurance.csv")
df.head()
```

```python
df = pd.get_dummies(df,["sex","smoker","region"],drop_first=True).astype(int)
df
```

```python
y = df["charges"]
x = df.drop("charges",axis = 1)
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
df_hata = pd.DataFrame()
df_hata["y"] = y
df_hata.head()
```

```python
y_tahmin = model.predict(x)
df_hata["error"]= y_tahmin
df_hata.head()
```

```python
mean_absolute_error(y,y_tahmin)
```

```python
mean_absolute_percentage_error(y,y_tahmin)
```

```python
mean_squared_error(y,y_tahmin)
```
